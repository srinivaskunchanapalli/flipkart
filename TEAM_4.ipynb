# **YOUTUBE VIDEO TO TEXT GENERATION**

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from youtube_transcript_api import YouTubeTranscriptApi

# Load pre-trained GPT-2 model and tokenizer
model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)

# Define the YouTube video ID
youtube_video_id = 'f5NJQiY9AuY'

# Fetch the video transcript using youtube_transcript_api
video_transcript = YouTubeTranscriptApi.get_transcript(youtube_video_id)

# Combine the transcript into a single string
text_input = ' '.join([item['text'] for item in video_transcript])

# Tokenize the input text
input_ids = tokenizer.encode(text_input, return_tensors='pt')

# Truncate the input text if it exceeds the model's maximum token limit
max_token_limit = model.config.max_position_embeddings
if input_ids.size(1) > max_token_limit:
    input_ids = input_ids[:, :max_token_limit]

# Generate text using the GPT-2 model
output = model.generate(input_ids, max_length=200, num_return_sequences=1, no_repeat_ngram_size=2, top_k=50, top_p=0.95)

# Decode the generated text
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

# Save the generated text to a file
with open('generated_text.txt', 'w', encoding='utf-8') as text_file:
    text_file.write(generated_text)

# Print confirmation message
print("Generated text saved in 'generated_text.txt'")
```

# **TEXT TO AUDIO GENERATION** (audio saved in a file)

```python
from gtts import gTTS
import os

# Path to the generated text file
generated_text_file = "generated_text.txt"

# Read the generated text from the file
with open(generated_text_file, "r") as file:
    generated_text = file.read()

# Convert text to speech using gTTS
tts = gTTS(text=generated_text, lang='en')

# Save the generated speech as an audio file
output_audio_file = "generated_audio_output.mp3"
tts.save(output_audio_file)

# Print confirmation message
print(f"Generated audio saved to {output_audio_file}")
```

# **FORMATTING TEXT IN IMAGE**

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from youtube_transcript_api import YouTubeTranscriptApi
from gtts import gTTS
from PIL import Image, ImageDraw, ImageFont
import os

# Save the generated text to an image
output_image_path = "generated_text_output.png"

# Create an image with white background
img = Image.new('RGB', (800, 600), color='white')
draw = ImageDraw.Draw(img)

# Use a default font if "arial.ttf" is not available
try:
    font = ImageFont.truetype("arial.ttf", size=20)
except IOError:
    font = ImageFont.load_default()

# Specify the position and wrap text
position = (10, 10)
max_width = 780

# Define function to wrap text at the edge
def wrap_text(text, width):
    lines = []
    words = text.split()
    current_line = words[0]
    for word in words[1:]:
        if font.getsize(current_line + ' ' + word)[0] <= width:
            current_line += ' ' + word
        else:
            lines.append(current_line)
            current_line = word
    lines.append(current_line)
    return lines

# Wrap text and draw onto the image with different font colors
wrapped_lines = wrap_text(generated_text, max_width)
text_color = ['black', 'red', 'blue', 'green', 'purple']  # You can customize the colors
for i, line in enumerate(wrapped_lines):
    draw.text(position, line, font=font, fill=text_color[i % len(text_color)])
    position = (position[0], position[1] + font.getsize(line)[1])
# img.show()
# Save the image
img.save(output_image_path)
# Print confirmation message
print(f"Generated text saved as an image: {output_image_path}")
```

# **TEXT TRANSLATION**

```python
!pip install translate
from transformers import GPT2LMHeadModel, GPT2Tokenizer
from youtube_transcript_api import YouTubeTranscriptApi
from translate import Translator


# Ask the user for the target language
target_language = input("Enter the target language code (e.g., 'te' for Telugu): ")

# Split the text into chunks of 500 characters or less
chunk_size = 500
text_chunks = [text_input[i:i + chunk_size] for i in range(0, len(text_input), chunk_size)]

# Initialize an empty list to store translated chunks
translated_text_chunks = []

# Translate each chunk and append to the list
for chunk in text_chunks:
    # Translate the chunk into the specified language
    translator = Translator(to_lang=target_language)
    translated_chunk = translator.translate(chunk)
    translated_text_chunks.append(translated_chunk)

# Concatenate the translated text chunks
translated_text = ' '.join(translated_text_chunks)

# Print or save the translated text
print("Original Text:")
print(text_input)
print("\nTranslated Text:")
print(translated_text)
```
